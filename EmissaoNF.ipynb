```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import os


options = webdriver.ChromeOptions()
options.add_experimental_option("prefs", {
  "download.default_directory": r"C:\Users\danie\Downloads",
  "download.prompt_for_download": False,
  "download.directory_upgrade": True,
  "safebrowsing.enabled": True
})

navegador = webdriver.Chrome(options=options)

# Passo 1:
# entrar na página de login

caminho = os.getcwd()
arquivo = caminho + r"\login.html"
navegador.get(arquivo)

# Passo 2:
# preencher o login e a senha
navegador.find_element(By.XPATH, '/html/body/div/form/input[1]').send_keys('***SEU EMAIL***')
navegador.find_element(By.XPATH, '/html/body/div/form/input[2]').send_keys('***SUA SENHA***')


# clicar no botão de fazer login
navegador.find_element(By.XPATH, '/html/body/div/form/button').click()

#Passo 3:
# preencher os dados da Nota Fiscal

# nome/razao social
navegador.find_element(By.NAME, 'nome').send_keys('Vitor')

# endereço 
navegador.find_element(By.NAME, 'endereco').send_keys('Rua Qualquer Rua')

# bairro 
navegador.find_element(By.NAME, 'bairro').send_keys('Centro')

# municipio
navegador.find_element(By.NAME, 'municipio').send_keys('Municipio Qualquer')

# cep
navegador.find_element(By.NAME, 'cep').send_keys('99999-999')

# uf
navegador.find_element(By.NAME, 'uf').send_keys('rn')
    
# cnpj/cpf
navegador.find_element(By.NAME, 'cnpj').send_keys('000.000.000-00')
    
# inscrição estadual
navegador.find_element(By.NAME, 'inscricao').send_keys('8238')
    
# descricao do produto/serviço
navegador.find_element(By.NAME, 'descricao').send_keys('Produto Hashtag Treinamentos Python')
    
# quantidade
navegador.find_element(By.NAME, 'quantidade').send_keys('92492')
    
#valor unitario
navegador.find_element(By.NAME, 'valor_unitario').send_keys('328327')
    
#valor total
navegador.find_element(By.NAME, 'total').send_keys('291919')

    
# Passo 4:
# clicar em emitir a Nota Fiscal
navegador.find_element(By.CLASS_NAME, 'registerbtn').click()
                            

# recarregar a página para limpar o formulario
navegador.refresh()
        
navegador.quit()

```
